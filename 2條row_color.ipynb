```python
import seaborn as sns; sns.set(color_codes=True)
import string
import pandas as pd
import numpy as np
```

```python
iris = sns.load_dataset("iris")
iris

samples = np.repeat(list(string.ascii_letters[0:8]),20)[:150] #將150個sample每20個分為1組
sample_cols = dict(zip(set(samples), sns.color_palette("cubehelix", 8))) #smple row_color

species = iris.pop("species") #分割species column到species參數
lut = dict(zip(species.unique(), "rby"))
row_colors = pd.DataFrame({'species':species.map(lut),
                          'sample':[sample_cols[i] for i in samples]}) #species row_color

g = sns.clustermap(iris, row_colors=row_colors,row_cluster=False)
```
